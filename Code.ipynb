```python
import pandas as pd
import seaborn as sns
import pandas, seaborn
import numpy as np
import matplotlib.pyplot as plt
from functools import reduce
```

```python
file1 = "Resources/Michaels.xlsx"
df = pd.read_excel(file1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
table=pd.pivot_table(df, values=['Deaths','Population'], index=['State','Year'],  aggfunc=np.sum)
table
```

```python
table.to_csv("Resources/table.csv", index=True)
```

```python
df.describe()
```

```python
sns.relplot(x='Deaths',y='Year', data=df)
fig = plt.gcf()
fig.set_size_inches(12, 10)
```

```python
ax = sns.pairplot(df)
```

```python
g= sns.relplot(x='Year',y='Deaths', kind='line', data=df)
fig = plt.gcf()
fig.set_size_inches(12, 10)
g.set_xticklabels(rotation=90)
```

```python
g= sns.factorplot(x='Year',y='Deaths', data=df)
fig = plt.gcf()
fig.set_size_inches(12, 10)
g.set_xticklabels(rotation=90)
```

```python
g= sns.catplot(x='Year',y='Dth_Rate', data=df)
fig = plt.gcf()
fig.set_size_inches(12, 10)
g.set_xticklabels(rotation=90)
```
