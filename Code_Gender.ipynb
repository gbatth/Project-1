```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib
from functools import reduce
```

```python
file1 = "Resources/Underlying Cause of Death, 1999_2018_Gender.xlsx"
df = pd.read_excel(file1)
```

```python
df.head()
```

```python
table=pd.pivot_table(df, values=['Deaths','Population'], index=['State','Year'], columns=['Gender'], aggfunc=np.sum)
table
```

```python
table.to_csv("Resources/table_gender.csv", index=True)
```

```python
g=sns.relplot(x='Year',y='Crude Rate', data=df, kind="line",col='Gender')
fig = plt.gcf()
fig.set_size_inches(20, 10)
g.set_xticklabels(rotation=90)
```

```python
ax = sns.pairplot(df)
```

```python
g= sns.relplot(x='Year',y='Crude Rate', kind='line', data=df, col='State')
fig = plt.gcf()
fig.set_size_inches(20, 10)
g.set_xticklabels(rotation=90)
```

```python
g= sns.factorplot(x='State',y='Crude Rate', data=df)
fig = plt.gcf()
fig.set_size_inches(12, 10)
g.set_xticklabels(rotation=90)
```

```python
g= sns.catplot(x='State',y='Rate', hue="Gender", data=df, kind="swarm")
fig = plt.gcf()
fig.set_size_inches(20, 10)
g.set_xticklabels(rotation=90)
```
